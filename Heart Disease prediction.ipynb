### Dummies using pandas to convert categorical value to one-hot encoding

```python
df=pd.get_dummies(df,columns=["cp","restecg"])
```

```python
df=df.rename(columns={"num       ":"target"})
```

```python
# normalizatoion rescaling purpose  change the values of numeric columns in the dataset to use a common scale.
numerical_cols=["age","trestbps","chol","thalach","oldpeak"]
cat_cols=list(set(df.columns)-set(numerical_cols)-{"target"})
```

```python
df.head()
```

#### splitting data

```python
df_train,df_test=train_test_split(df,test_size=0.2,random_state=42)
```

```python
len(df_train)
```

```python
len(df_test)
```

```python
# scaling part 1300 as outlier 
scaler=StandardScaler()

def get_feature_and_target_arrays(df,numerical_cols,cat_cols,scaler):
    x_numeric_scaled=scaler.fit_transform(df[numerical_cols])# 0 to 1 numerical col
    x_categorical=df[cat_cols].to_numpy()# categorical cols
    x=np.hstack((x_categorical,x_numeric_scaled))
    y=df["target"]
    
    return x,y
    # np.hstack(a,b) single array 
# a=[1,2,3]
# b=[2,5,6]
# c=[1,2,3,4,5,5]
```

```python
x_train,y_train =get_feature_and_target_arrays(df_train,numerical_cols,cat_cols,scaler)
```

### Train

```python
clf=LogisticRegression()
clf.fit(x_train,y_train)
```

```python
x_test,y_test =get_feature_and_target_arrays(df_test,numerical_cols,cat_cols,scaler)
```

```python
#The predict method predicts our actual testing dataset with regard to the fitted (training) data.
test_pred=clf.predict(x_test)
test_pred.mean()
```

```python
mean_squared_error(y_test,test_pred)
```

```python
accuracy_score(y_test,test_pred)
```

```python
confusion_matrix(y_test ,test_pred)
```

```python
# Decision Tree classifier - giving poor accuracy
dc_clf=DecisionTreeClassifier()
dc_clf.fit(x_train,y_train)

dlf_predict = dc_clf.predict(x_test)

print(mean_squared_error(y_test,dlf_predict))
print(accuracy_score(y_test,dlf_predict))
```

```python
# Random Forest Classifier
rc_clf=RandomForestClassifier()
rc_clf.fit(x_train,y_train)

rc_predict = rc_clf.predict(x_test)

print(mean_squared_error(y_test,rc_predict))
print(accuracy_score(y_test,rc_predict))
```

```python
# SVM
svm_clf=SVC()
svm_clf.fit(x_train,y_train)

svm_predict = svm_clf.predict(x_test)

print(mean_squared_error(y_test,svm_predict))
print(accuracy_score(y_test,svm_predict))
```
